```python


replica2 = df.copy(deep = True)
replica2
```

```python
replica2["LanguageWorkedWith"]
```

```python
#removing all NAN values from LanguageWorkedWith column
replica2 = replica2[replica2["LanguageWorkedWith"].isna() != 1]
replica2["LanguageWorkedWith"]
```

```python
# Fitering only developers
replica2 = replica2[replica2["MainBranch"]=="I am a developer by profession"]
replica2["knows_python"] = replica2["LanguageWorkedWith"].str.contains("python",case=False, na= False)
replica2
```

```python
developers = replica2[["Country","knows_python"]]

#filtering developers who know python
developers_knows_python = developers[developers["knows_python"] == 1]

#grouping all developers by country
developers = developers.groupby(["Country"]).count()

#grouping developers who knows python
developers_knows_python = developers_knows_python.groupby(["Country"]).count()


#percentage in every country
percent_of_pythonDeveloper_in_every_country = developers_knows_python/developers * 100

# final output
percent_of_pythonDeveloper_in_every_country

```

# 3.  Generate the report for the average salary of developers based on the continent.


```python


replica3 = df.copy(deep = True)
replica3
```

```python
replica3["Country"].unique()
```

```python
replica3 = replica3[replica3["Country"] != "Other Country (Not Listed Above)"]

#filtring developer
replica3 = replica3[replica3["MainBranch"] == "I am a developer by profession"]

#Adding continent Coloumn
continent = pd.read_csv("./content/Countries-Continents.csv")

#checking for contradictories name
diff_country = []
for country in replica3['Country']:
    for country1 in continent['Country']:
        if country not in country1:
            diff_country.append(country)
            break
    break
diff_country

```

```python
replica3.replace('United States','US',inplace=True)

#merge continent in df

replica3 = replica3.merge(continent, how="left", on = ["Country"])

data_developers = replica3[["ConvertedComp","Continent"]]
avg_salary = data_developers.groupby(["Continent"]).mean()

avg_salary = avg_salary.rename(columns={'ConvertedComp':'Avg_salary_in_USD'})
avg_salary
```

# 4.  Based on this survey, what will be the most desired programming language for the year  2020?



```python
replica4 = df.copy(deep = True)
replica4
```

```python
# removing NAN from desired Skills
replica4 = replica4[replica4["LanguageDesireNextYear"].isna() != 1]

#spliting all skills and appending in list
list = []
for l in replica4["LanguageDesireNextYear"]:
    x = l.split(";")
    for i in x:
        list.append(i)
        
lang = pd.Series(list)

#final output
lang.value_counts()
```

#  5.Generate the report for the people who code as a hobby based on their gender and continent.





```python
replica5 = replica3.copy(deep = True)
replica5
```

```python
replica5["Gender"].unique()
```

```python
def gender(row):
    if row == "Man":
        return "Man"
    if row == "Woman":
        return "Woman"
    else:
        return "Others"
    
replica5["Gender"] = replica5["Gender"].apply(gender)
```

```python
hobby_info = replica5.groupby(["Gender","Continent"]).sum()
hobby_info
```


# 6.Generate the report for job and career satisfaction of developers based on their gender and continent.



```python
replica6 = replica5.copy(deep = True)
replica6
```

```python
# removing all NAN values from job satisfaction column
replica6 = replica6[replica6["JobSat"].isna() != 1]
replica6["JobSat"]
```

```python
replica6["Job_very_satisfied"] = replica6['JobSat'] == "Very satisfied"
replica6["Job_very_dissatisfied"] = replica6['JobSat'] == "Very dissatisfied"
replica6["Job_Slightly_satisfied"] = replica6['JobSat'] == "Slightly satisfied"
replica6["Job_very_dissatisfied"] = replica6['JobSat'] == "Slightly dissatisfied"
replica6["Job_Neither_satisfied_nor_dissatisfied"] = replica6['JobSat'] == "Neither satisfied nor dissatisfied"

job_satisfaction_info = replica6.groupby(["Gender" , "Continent"])
job_satisfaction_info.sum()


```
